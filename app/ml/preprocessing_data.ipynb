# Remove Formula for to make it easier to read

```python
import re

def remove_hypen(abstract_series:pd.Series):
    return abstract_series.apply(lambda text: re.sub(r'[^\w\s]', '', str(text)))
```

```python
datas['abstract']= remove_hypen(datas['abstract'])
```

```python
datas['abstract']
```

```python
datas
```

```python
datas['preprocessed_absract'] = datas['abstract']
datas
```

```python
from nltk.tokenize import word_tokenize



datas['preprocessed_absract']=datas['preprocessed_absract'].apply(lambda x:word_tokenize(x))
```

```python
datas
```

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

# abstracts are tokenized (lists of tokens); stem each token instead of trying to stem the list
# datas['preprocessed_absract'] = datas['preprocessed_abstract'].apply(lambda tokens: [stemmer.stem(t) for t in tokens])
datas['preprocessed_absract']=datas['preprocessed_absract'].apply(lambda x:[stemmer.stem(t)for t in x])
```
